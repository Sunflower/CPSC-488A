```python
concreteness_ratings = pd.read_csv('data/Concreteness_ratings_Brysbaert_et_al_BRM.txt', sep='\t')
display(len(concreteness_ratings))
concreteness_ratings.head()
```

```python
words_with_ac_ratings = (
    all_words.merge(concreteness_ratings[['Word', 'Conc.M']],
                    how='left',
                    left_on='lemma',
                    right_on='Word')
    .sort_values(by='word_id')
    .fillna(-1)[['word_id', 'lemma', 'Conc.M']]
    .rename(columns={'Conc.M':'ac'})
)
display(len(words_with_ac_ratings))
words_with_ac_ratings.sample(10)
```

```python
words_with_ac_ratings.to_csv('data/words_with_ac_ratings.csv', index=False)
print('Success!')
```

```python
words_with_deps = pd.read_csv('data/words_with_deps.csv')
display(len(words_with_deps))
words_with_deps.head(50)
```

```python
words_with_ac_ratings_deps = words_with_ac_ratings.merge(
    words_with_deps,
    left_on='word_id',
    right_on='word_id'
).set_index('word_id')

display(len(words_with_ac_ratings_deps))
words_with_ac_ratings_deps.head(30)
```

```python
words_with_ac_ratings_deps.columns
```

Itertuples renames columns with spaces or unaccepted symbols.

```python
words_with_ac_ratings_deps.columns = words_with_ac_ratings_deps.columns.str.replace(r':', '_')
```

Build dictionary of types of dependencies.

```python
dep_dict_empty = {
    'amod': 0,
    'nsubj': 0,
    'obj': 0,
    'appos': 0,
    'flat': 0,
    'aux': 0,
    'acl_relcl': 0,
    'det': 0,
    'nummod': 0,
    'compound': 0,
    'case': 0,
    'nmod': 0,
    'mark': 0,
    'acl': 0,
    'obl': 0,
    'ccomp': 0,
    'nmod_poss': 0,
    'cc': 0,
    'conj': 0,
    'punct': 0,
    'advmod': 0,
    'nsubj_pass': 0,
    'aux_pass': 0,
    'cop': 0,
    'parataxis': 0,
    'advcl': 0,
    'compound_prt': 0,
    'expl': 0,
    'obl_tmod': 0,
    'det_predet': 0,
    'csubj': 0,
    'xcomp': 0,
    'iobj': 0,
    'discourse': 0,
    'list': 0,
    'vocative': 0,
    'nmod_tmod': 0,
    'obl_npmod': 0,
    'nmod_npmod': 0,
    'cc_preconj': 0,
    'fixed': 0,
    'goeswith': 0,
    'reparandum': 0
}
```

```python
ast.literal_eval(words_with_ac_ratings_deps.loc[0, 'amod'])
```

```python
import copy

dep_dicts = []
for row in words_with_ac_ratings_deps.itertuples():
    dep_dict = copy.deepcopy(dep_dict_empty)
    for dep in dep_dict.keys():
        dep_indices = ast.literal_eval(getattr(row, dep))
        if not len(dep_indices):
            continue
            
        sum_ratings = 0
        for i in dep_indices: 
            rating = words_with_ac_ratings_deps.loc[i, 'ac']
            sum_ratings += rating
            
        dep_dict[dep] = sum_ratings/len(dep_indices)
        
    dep_dict['word_id'] = row.Index
    dep_dicts.append(dep_dict)
    if row.Index % 1000 == 0:
        print(row.Index)
    
print(len(dep_dicts))
```
